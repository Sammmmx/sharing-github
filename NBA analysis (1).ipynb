```python
def plot(Players):
    for name in Players:
        mlp.plot(Salary[Pdict[name]], c='Red', marker='o', ms=7, label=name)
    mlp.legend(loc='lower right')
    mlp.xticks(list(range(0,10)), Seasons,)
    mlp.show()
    mlp.rcParams['figure.figsize'] = 12,8
```

```python
def plot(Players):
    col = {'KobeBryant':'Red', 'CarmeloAnthony':'Pink', 'LeBronJames':'Blue'}
    Mark = {'KobeBryant':'o', 'CarmeloAnthony':'s', 'LeBronJames':'D'}
    for name in Players:
        mlp.plot(Salary[Pdict[name]], c=col[name], marker=Mark[name], ms=7, label=name)
    mlp.legend(loc='lower right')
    mlp.xticks(list(range(0,10)), Seasons,)
    mlp.show()
    mlp.rcParams['figure.figsize'] = 12,8
```

```python
plot(['KobeBryant', 'CarmeloAnthony', 'LeBronJames'])
```

```python
def plot(data, Playerlist=Players):
    col = {'KobeBryant':'Red', 'CarmeloAnthony':'Pink', 'LeBronJames':'Blue'}
    Mark = {'KobeBryant':'o', 'CarmeloAnthony':'s', 'LeBronJames':'D'}
    for name in Playerlist:
        mlp.plot(data[Pdict[name]], c=col[name], marker=Mark[name], ms=7, label=name)
    mlp.legend(loc='lower right')
    mlp.xticks(list(range(0,10)), Seasons,)
    mlp.show()
    mlp.rcParams['figure.figsize'] = 12,8
```
